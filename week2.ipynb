## 1. Lists ‚Üí Ordered, Changeable
- Think of them as shopping lists üìã.

```python
fruits = ["apple", "banana", "cherry"]
print(fruits[0])        # apple
fruits.append("orange") # add
fruits.remove("banana") # remove
print(fruits)           # ['apple', 'cherry', 'orange']s
```

## 2. Tuples ‚Üí Ordered, Unchangeable
- Like a list, but locked üîí.

```python
coordinates = (10, 20)
print(coordinates[0])   # 10
# coordinates[0] = 30 ‚ùå Error (cannot modify)
```

## 3. Dictionaries ‚Üí Key:Value pairs
- Think of them as contact books üìñ.

```python
person = {"name": "Roy", "age": 25, "color": "Black"}
print(person["name"])     # Roy
person["age"] = 26        # update
person["hobby"] = "coding" # add new
print(person)
```

## 4. Sets ‚Üí Unique values, no duplicates
- Like a bag of unique items üéí.

```python
numbers = {1, 2, 2, 3, 4}
print(numbers) # {1, 2, 3, 4} (no duplicates)

a = {1, 2, 3}
b = {3, 4, 5}
print(a | b)  # union ‚Üí {1, 2, 3, 4, 5}
print(a & b)  # intersection ‚Üí {3}
```

## üëâ Project: Student Gradebook
1. Use a list of dictionaries to store student info (name, age, grade).
2. Add a new student.
3. Remove a student.
4. Print all students.
5. (Bonus) Find the average grade.

```python
students = []

while True:
    num = input('Enter 1=Add, 2=Remove, 3=Print, 4=Exit, 5=Average: ')

    if num == '1':
        name = input("Name: ")
        age = int(input("Age: "))
        grade = float(input("Grade: "))
        students.append({"name": name, "age": age, "grade": grade})
        print(f"Added {name}!")

    elif num == '2':
        name = input("Name to remove: ")
        found = False
        for student in students:
            if student["name"] == name:
                students.remove(student)
                print(f"Removed {name}!")
                found = True
                break
        if not found:
            print("Student not found.")

    elif num == '3':
        if students:
            print("--- Student List ---")
            for s in students:
                print(f"Name: {s['name']}, Age: {s['age']}, Grade: {s['grade']}")
        else:
            print("No students yet.")

    elif num == '4':
        print("Goodbye!")
        break

    elif num == '5':
        if students:
            avg = sum(s["grade"] for s in students) / len(students)
            print(f"Average Grade: {avg:.2f}")
        else:
            print("No students yet.")

    else:
        print("Invalid choice, try again.")
```

## üéØ Week 2 Final Practice Project (No CSV yet)

### üëâ Contact Book
- Store contacts as a list of dictionaries (each dictionary has: name, phone, email).
- Menu options:
  1. Add a contact
  2. Remove a contact
  3. Find a contact by name
  4. List all contacts
  5. Exit

```python
# contacts list
contacts = []

def find_index_by_name(name):
    """Return contact index by name (case-insensitive), or -1 if not found."""
    name = name.strip().lower()
    for i, c in enumerate(contacts):
        if c["name"].strip().lower() == name:
            return i
    return -1

while True:
    num = input('\nEnter 1=Add, 2=Remove, 3=Find, 4=List, 5=Edit, 6=Exit: ').strip()

    # add
    if num == '1':
        name = input("Name: ").strip()
        if find_index_by_name(name) != -1:
            print("That name already exists. Try Edit instead.")
            continue
        phone = input("Phone: ").strip()
        email = input("Email: ").strip()
        contacts.append({"name": name, "phone": phone, "email": email})
        print(f"Added {name}!")

    # remove
    elif num == '2':
        name = input("Name to remove: ").strip()
        idx = find_index_by_name(name)
        if idx == -1:
            print("Contact not found.")
        else:
            removed = contacts.pop(idx)
            print(f"Removed {removed['name']}!")

    # find
    elif num == '3':
        name = input("Name to find: ").strip()
        idx = find_index_by_name(name)
        if idx == -1:
            print("Contact not found.")
        else:
            c = contacts[idx]
            print(f"Name: {c['name']}, Phone: {c['phone']}, Email: {c['email']}")

    # list
    elif num == '4':
        if contacts:
            print("\n--- Contacts ---")
            for i, c in enumerate(contacts, 1):
                print(f"{i}. {c['name']} | {c['phone']} | {c['email']}")
        else:
            print("No contacts yet.")

    # edit (optional)
    elif num == '5':
        name = input("Name to edit: ").strip()
        idx = find_index_by_name(name)
        if idx == -1:
            print("Contact not found.")
            continue
        c = contacts[idx]
        new_name  = input(f"New name  (Enter to keep: {c['name']}): ").strip()
        new_phone = input(f"New phone (Enter to keep: {c['phone']}): ").strip()
        new_email = input(f"New email (Enter to keep: {c['email']}): ").strip()

        if new_name:  c["name"] = new_name
        if new_phone: c["phone"] = new_phone
        if new_email: c["email"] = new_email
        print("Contact updated!")

    # exit
    elif num == '6':
        print("Goodbye!")
        break

    else:
        print("Invalid choice, try again.")
```
