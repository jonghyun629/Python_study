## üå°Ô∏èProject: Temperature Converter

Your program should:
- Ask the user to type a temperature in Celsius.
- Convert it to Fahrenheit.
- Print the result nicely.
- (Bonus) Do the reverse: Fahrenheit ‚Üí Celsius.

```python
# Ask temperature in Celsius or Fahrenheit
choice = input("Convert from (C)elsius or (F)ahrenheit? ").lower()

# Convert temp
if choice == 'c':
    c = float(input("Enter Celsius: "))
    f = (c * 9/5) + 32
    print(f"{c:.2f}¬∞C = {f:.2f}¬∞F")
elif choice == 'f':
    f = float(input("Enter Fahrenheit: "))
    c = (f - 32) * 5/9
    print(f"{f:.2f}¬∞F = {c:.2f}¬∞C")
else:
    print("Invalid choice.")
```

## üé≤ Week 1 Final Challenge Project: Simple Calculator

### üìå Requirements

- Ask the user for two numbers.

- Ask the user to choose an operation: add, subtract, multiply, divide.

- Perform the calculation and print the result.

- (Bonus) Keep running until the user types "exit".

```python
def cal():
    while True:  # keep running until break
        # Asking the user for two numbers
        num1 = float(input('Enter a number: '))
        num2 = float(input('Enter another number: '))

        # Choosing an operation
        operation = int(input('Enter 1 for addition, 2 for subtraction, 3 for multiplication, 4 for division, 5 to exit: '))

        # Function
        if operation == 1:
            print(f"{num1} + {num2} = {num1 + num2}")
        elif operation == 2:
            print(f"{num1} - {num2} = {num1 - num2}")
        elif operation == 3:
            print(f"{num1} * {num2} = {num1 * num2}")
        elif operation == 4:
            if num2 == 0:
                print("Error: Cannot divide by zero!")
            else:
                print(f"{num1} / {num2} = {num1 / num2}")
        elif operation == 5:
            print("Exiting calculator... Goodbye!")
            break
        else:
            print("Invalid operation, please choose 1‚Äì5")

# Run it
cal()
```
