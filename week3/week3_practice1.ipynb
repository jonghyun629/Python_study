4. Use `enumerate()` to print a numbered shopping list:
  1. milk
  2. eggs
  3. bread

```python
shopping_list = ["milk", "eggs", "bread"]

for index, item in enumerate(shopping_list, start=1):
  print(f"{index}. {item}")
```
