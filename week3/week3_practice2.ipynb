1. Count how many times the letter "a" appears in `"bananas are amazing"`.

```python
word = "bananas are amazing"
letter = "a"
count = 0

for alpha in word:
  if alpha in letter:
    count += 1

print(count)
```

2. Search if `"apple"` exists in the list `["banana", "cherry", "apple", "grape"]`.

```python
fruits = ["banana", "cherry", "apple", "grape"]
fruit = 'apple'

if fruit in fruits:
  print("It exists")
else:
  print("It doesn't exist")
```

3. From `[12, 45, 67, 24, 90, 33]`, create a new list with only numbers greater than 50.

```python
l = [12, 45, 67, 24, 90, 33]

new = []
for i in l:
  if i > 50:
    new.append(i)

print(new)
```
