## üîë Day 1: for Loops, while Loops, range()

1. `for` **loop ‚Üí for fixed repetitions / iterating over data**

```python
fruits = ["apple", "banana", "cherry"]
for f in fruits:
    print(f)
```

2. `while` **loop ‚Üí for unknown repetitions (runs until condition is false)**

```python
count = 0
while count < 3:
    print("Hello")
    count += 1
```

3. `range()` **‚Üí number sequence**

```python
for i in range(5):
    print(i)
```

```python
for i in range(2, 10, 2):  # start=2, stop=10, step=2
    print(i)
```

4. `enumerate()` **‚Üí get index + value**

```python
fruits = ["apple", "banana", "cherry"]
for index, fruit in enumerate(fruits, start=1):
    print(index, fruit)
```

## üîë Day 2: Loop Patterns

1. **Counting items**

üëâ Example: Count vowels in a word

```python
word = "python programming"
vowels = "aeiou"
count = 0

for alpha in word:
    if alpha in vowels:
        count += 1

print("Number of vowels:", count)
```

2. **Searching**

üëâ Example: Check if a number is in a list

```python
numbers = [10, 20, 30, 40, 50]
target = 30
found = False

for n in numbers:
    if n == target:
        found = True
        break

print("Found:", found)
```

3. **Filtering**

üëâ Example: Keep only even numbers from a list

```python
numbers = [1, 2, 3, 4, 5, 6]
evens = []

for n in numbers:
    if n % 2 == 0:
        evens.append(n)

print("Even numbers:", evens)
```

## üîë Day 3 Concepts

1. `break` ‚Üí exit the loop immediately

```python
for i in range(1, 6):
    if i == 3:
        break
    print(i)
```

2. `continue` ‚Üí skip this iteration, move to the next

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

3. Nested loops ‚Üí a loop inside another loop

üëâ Example: Multiplication table

```python
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i} x {j} = {i*j}")
```

# üîë Day 4 What is a List Comprehension?

A **list comprehension** is a compact way to create lists using a single line of code.

### General Form:

```python
new_list = [expression for item in iterable if condition]
```

### 1. Basic Example

**Normal loop:**

```python
numbers = [1, 2, 3, 4, 5]
squares = []
for n in numbers:
    squares.append(n * n)
print(squares)
```

**With comprehension:**

```python
squares = [n * n for n in numbers]
print(squares)
```

### 2. With Condition

üëâ Only keep even numbers:

```python
numbers = [1, 2, 3, 4, 5, 6]
evens = [n for n in numbers if n % 2 == 0]
print(evens)
```

### 3. With Strings

üëâ Convert words to uppercase:

```python
words = ["python", "is", "fun"]
uppercased = [w.upper() for w in words]
print(uppercased)
```

### 4. Nested List Comprehension

üëâ Flatten a 2D list:

```python
matrix = [[1, 2], [3, 4], [5, 6]]
flat = [num for row in matrix for num in row]
print(flat)
```

## üìù Project 1: Word Counter

üëâ Count how many times each word appears in a sentence.

# **Example**

```python
sentence = "python is fun and python is powerful"
```

Expected output:

```python
{'python': 2, 'is': 2, 'fun': 1, 'and': 1, 'powerful': 1}
```

üí° Hints:

- Use `.split()` to break the sentence into words.

- Use a dictionary to count frequencies.

```python
sentence = 'Enter a sentence'
word = sentence.split()
#print(word)

count = {}
for i in word:
  if i in count:
    count[i] += 1
  else:
    count[i] = 1

print(count)
```

## üìù Project 2: Student Scores Analyzer

üëâ Given a list of students and scores, do analysis.

```python
students = [
    {"name": "Alice", "score": 85},
    {"name": "Bob", "score": 72},
    {"name": "Charlie", "score": 65},
    {"name": "Diana", "score": 92},
]
```

Expected tasks:
1. Print all students who scored above 70.
2. Calculate the average score.
3. Find the top student (highest score).

```python
students = [
    {"name": "Alice", "score": 85},
    {"name": "Bob", "score": 72},
    {"name": "Charlie", "score": 65},
    {"name": "Diana", "score": 92},
]

print('---students who scored above 70---')
for student in students:
  #print(student)
  if student["score"] > 70:
    print(student["name"])

print()
print('---average score---')
total = 0
for student in students:
  total += student['score']
  #print(total)
  avg = total / len(students)
print(avg)

print()
print('---top student---')
top = students[0]
for student in students:
  if student['score'] > top['score']:
    top = student
print(top)
```

---

üí° Hints:

- A number is prime if it‚Äôs only divisible by 1 and itself.
- Use nested loops to check divisibility.
- Optimize: skip even numbers except 2.

```python
prime = [ ]
# numbers up to 100
for i in range(2,101):
  prime.append(i)

  for j in range(2,i):
    if i % j == 0:
      prime.remove(i)
      break

print(prime)
```
