1. Use a loop with `break`: print numbers 1 ‚Üí 10, but stop if you reach 7.

```python
for i in range(1, 11):
  if i == 7:
    break
  print(i)
  i =+ 1
```

2. Use a loop with `continue`: print numbers 1 ‚Üí 10, but skip 5.

```python
for i in range(1, 11):
  if i == 5:
    continue
  print(i)
  i =+ 1
```

3. Use a **nested loop** to print all pairs of numbers from 1‚Äì3:

(1,1) (1,2) (1,3)

(2,1) (2,2) (2,3)

(3,1) (3,2) (3,3)

```python
for i in range(1,4):
  for j in range(1,4):
    print(f"({i},{j})", end = " ")
    if j ==3:
      print()
```
