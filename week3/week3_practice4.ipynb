1. Create a list of squares from 1 ‚Üí 10 using a list comprehension.

```python
squares = [i**2 for i in range(1,11)]

print(squares)
```

2. Create a list of only the odd numbers from 1 ‚Üí 20.

```python
odd = [i for i in range(1,21) if i % 2 != 0]

print(odd)
```

3. Given `words = ["apple", "banana", "cherry"]`, make a new list with the **length of each word**.

(Expected: [5, 6, 6])

```python
words = ["apple", "banana", "cherry"]

len_words = [len(i) for i in words]

print(len_words)
```

4. Flatten this list with a comprehension:

`[[1,2,3], [4,5], [6,7,8]]`

```python
num = [[1,2,3], [4,5], [6,7,8]]

'''
for i in num:
  for j in i:
    print(j)
'''

flat = [j for i in num for j in i]
print(flat)
```
