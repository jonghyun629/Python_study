```python
# 1. Todo List Manager ‚Üí add/remove tasks, save/load from `todo.txt`.

# add a todo list with a tick box and a due date
def add_task(task, due_date):
    with open("todo.txt", "a") as f:
        f.write("[ ] " +  task + " (due: " + due_date + ")\n")

# remove the ticked todo
def remove_task(task):
    lines = []
    with open("todo.txt", "r") as f:
        lines = f.readlines()
    with open("todo.txt", "w") as f:
        for line in lines:
            if line.strip() != "[x] " + task:
                f.write(line)

# show todo and tick the todo
def show_tasks():
    with open("todo.txt", "r") as f:
        tasks = f.readlines()
    for i, task in enumerate(tasks):
        print(f"{i + 1}. {task.strip()}")

def tick_task(index):
    lines = []
    with open("todo.txt", "r") as f:
        lines = f.readlines()
    with open("todo.txt", "w") as f:
        for i, line in enumerate(lines):
            if i == index:
                f.write("[x] " + line[4:])
            else:
                f.write(line)

# save and load the todo
def save_tasks():
    with open("todo.txt", "r") as f:
        tasks = f.readlines()
    with open("todo.txt", "w") as f:
        for task in tasks:
            f.write(task)

def load_tasks():
    with open("todo.txt", "r") as f:
        tasks = f.readlines()
    for task in tasks:
        print(task.strip())

def main():
    while True:
        print("\nTodo List Manager")
        print("1. Add Task")
        print("2. Remove Task")
        print("3. Show Tasks")
        print("4. Tick Task")
        print("5. Save Tasks")
        print("6. Load Tasks")
        print("7. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            task = input("Enter task: ")
            due_date = input("Enter due date: ")
            add_task(task, due_date)
        elif choice == "2":
            task = input("Enter task to remove: ")
            remove_task(task)
        elif choice == "3":
            show_tasks()
        elif choice == "4":
            index = int(input("Enter task number to tick: ")) - 1
            tick_task(index)
        elif choice == "5":
            save_tasks()
        elif choice == "6":
            load_tasks()
        elif choice == "7":
            break
        else:
            print("Invalid choice. Please try again.")

main()
```

```python
# 2. Write a program that creates `grades.csv` with columns `Name, Score` and 3 students. Then read and print each row.

import csv

student1 = input('Enter the name of student 1: ')
score1 = input('Enter the score of student 1: ')

student2 = input('Enter the name of student 2: ')
score2 = input('Enter the score of student 2: ')

student3 = input('Enter the name of student 3: ')
score3 = input('Enter the score of student 3: ')

# Writing CSV
with open("students.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Name", "Score"])
    writer.writerow([student1, score1])
    writer.writerow([student2, score2])
    writer.writerow([student3, score3])

# Reading CSV
with open("students.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

```python
# 3. Save a Python dictionary to a `settings.json` file, then load it back.

import json

data = {
    'name':'Roy',
    'age':25,
    'is_student':True
}

# save to file
with open("settings.json", "w") as f:
    json.dump(data, f)

# load from file
with open("settings.json", "r") as f:
    loaded = json.load(f)

print(loaded)
```
