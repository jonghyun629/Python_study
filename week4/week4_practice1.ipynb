```python
# 1. Write a function square(n) that returns the square of a number.

def square(n):
  return n**2

num = int(input('Enter a number: '))
print(square(num))
```

```python
# 2. Write a function is_even(n) that returns True if a number is even, else False.

def is_even(n):
  if n % 2 == 0:
    return True
  else:
    return False

num = int(input('Enter a number: '))
print(is_even(num))
```

```python
# 3. Write a function greet(name) that prints "Hello, <name>!".

def greet(name):
  print(f'Hello!, {name}!')

name = input('Enter your name: ')
greet(name)
```

```python
'''
4. Write a function calculate(a, b, operation) that:
  - adds if operation == "add"
  - subtracts if "sub"
  - multiplies if "mul"
  - divides if "div"
  - else prints "Invalid operation"
'''

def calculate(a, b, operation):
  if operation == 'add':
    return a + b
  elif operation == 'sub':
    return a - b
  elif operation == 'mul':
    return a * b
  elif operation == 'div':
    return a / b
  else:
    print('Invalid operation')

a = int(input('Enter a number: '))
b = int(input('Enter a number: '))
operation = input('Enter an operation: add, sub, mul, div: ')
print(calculate(a, b, operation))
```
