```python
# 1. Write a function multiply(a, b) that returns the product of two numbers.

def multiply(a, b):
  return a*b

a = int(input('Enter a number: '))
b = int(input('Enter a number: '))

multiply(a, b)
```

```python
# 2. Write a function square_and_cube(n) that returns both the square and cube of a number.

def square_and_cube(n):
  return n**2, n**3

num = int(input('Enter a number: '))
print(square_and_cube(num))
```

```python
# 3. Show the difference between a local and global variable with the same name.

outside = 'Roy'

def inside():
  name = 'Roy'
  return name

print(outside)
print(inside())
```

```python
'''
4. (Challenge üí°) Write a function circle_stats(radius) that returns both area and circumference of a circle (œÄr¬≤ and 2œÄr).
'''

def circle_stats(radius):
  return 3.14*radius**2, 2*3.14*radius

radius = int(input('Enter a radius: '))
print(circle_stats(radius))
```
